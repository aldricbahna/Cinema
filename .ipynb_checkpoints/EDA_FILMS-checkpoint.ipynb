## Décennies

```python
def decennie(annee):
    if annee<1930:
        return "20's"
    elif annee<1940:
        return "30's"
    elif annee<1950:
        return "40's"
    elif annee<1960:
        return "50's"
    elif annee<1970:
        return "60's"
    elif annee<1980:
        return "70's"
    elif annee<1990:
        return "80's"
    elif annee<2000:
        return "90's"
    elif annee<2010:
        return "2000's"
    elif annee<2020:
        return "2010's"
    else:
        return "2020's"
        
df['Décennie']=df['Sortie'].map(decennie)
df['Décennie']
```

```python
df_decennies=df.groupby('Décennie').agg(
    Nombre=('Nom','nunique'),
    Note_moyenne_perso=('Note','mean'),
    Ecart_type=('Note','std'),
    Note_moyenne_Eugenie=('Eugénie','mean'),
    Note_moyenne_Letterbox=('Letterbox','mean'),
    Note_moyenne_SensCritique=('Sens Critique','mean')).reset_index()
order=["2020's", "2010's", "2000's", "90's", "80's", "70's", "60's", "50's","30's"]
df_decennies['Décennie'] = pd.Categorical(df_decennies['Décennie'], categories=order, ordered=True)
df_decennies = df_decennies.sort_values(by='Décennie').reset_index(drop=True)
df_decennies
```

```python
df_melted = pd.melt(df_decennies, id_vars='Décennie', value_vars=['Note_moyenne_perso', 'Note_moyenne_Eugenie'],
                    var_name='Type', value_name='Moyenne')
plt.figure(figsize=(10, 6))
sns.barplot(data=df_melted, x='Décennie', y='Moyenne', hue='Type', palette="dark", alpha=.6)
plt.legend()
plt.title("Comparaison de nos moyennes de films en fonction de la décennie")
plt.xticks(rotation=45);
```

```python
def pays_principal(pays):
    if pays.startswith('France'):
        return "France"
    elif pays.startswith('Etats-Unis'):
        return "Etats-Unis"
    elif pays.startswith('Japon'):
        return "Japon"
    elif pays.startswith('Corée du Sud'):
        return "Corée du Sud"
    elif pays.startswith('Hong Kong'):
        return "Hong Kong"
    elif pays.startswith('Royaume-Uni'):
        return "Royaume-Uni"
    elif pays.startswith('Italie'):
        return "Italie"
    elif pays.startswith('Canada'):
        return "Canada"
    elif pays.startswith('Allemagne'):
        return "Allemagne"   
```

```python
df['Pays principal'] = df['Pays'].map(pays_principal)
df_pays=df.groupby('Pays principal').agg(
    Nombre=('Nom','nunique'),
    Note_moyenne_perso=('Note','mean'),
    Ecart_type_perso=('Note','std'),
    Note_moyenne_Eugenie=('Eugénie','mean'))
df_pays
```

```python
acteurs_series = df['Acteurs'].fillna('').str.split(',')

acteurs_flat = [acteur.strip() for sous_liste in acteurs_series for acteur in sous_liste if acteur]


acteur_counts = pd.Series(acteurs_flat).value_counts()
acteur_counts_df=acteur_counts.reset_index()
acteur_counts_df.columns=['Acteur','Nb_films']
acteur_counts_df
```

```python
df['Acteurs'] = df['Acteurs'].fillna('').str.split(',')

# Fonction pour calculer la moyenne des notes pour un acteur
def moyenne_notes(acteur, df):
    # Concatène les notes des films dans lesquels l'acteur apparaît
    notes = df[df['Acteurs'].apply(lambda x: acteur in [a.strip() for a in x])]['Note']
    if not notes.empty:
        return notes.mean()
    else:
        return None

# Appliquer la fonction pour chaque acteur dans acteur_counts_df
acteur_counts_df['Moyenne_Notes'] = acteur_counts_df['Acteur'].apply(lambda x: moyenne_notes(x, df))

# Afficher le DataFrame final
print(acteur_counts_df)
```

```python
import plotly.graph_objects as go
from math import pi

# Préparation des données
categories = dfc_genres.columns
num_vars = len(categories)

# Ajout de la première catégorie à la fin pour fermer le graphique
angles = [n / float(num_vars) * 2 * pi for n in range(num_vars)]
angles += angles[:1]

# Données pour Aldric
values_note = dfc_genres.loc['Aldric'].values.tolist()
values_note += values_note[:1]  # Fermer la boucle

# Données pour Jules
values_jules = dfc_genres.loc['Jules'].values.tolist()
values_jules += values_jules[:1]  # Fermer la boucle

# Création du radar chart avec Plotly
fig = go.Figure()

# Tracé pour Aldric
fig.add_trace(go.Scatterpolar(
    r=values_note,  # Valeurs
    theta=categories.tolist() + [categories[0]],  # Catégories avec la première répétée
    fill='toself',  # Remplissage
    name='Aldric',
    line=dict(color='#1A237E'),
    mode='lines'  # Supprime les gros points
))

# Tracé pour Jules
fig.add_trace(go.Scatterpolar(
    r=values_jules,
    theta=categories.tolist() + [categories[0]],
    fill='toself',
    name='Jules',
    line=dict(color='red'),
    mode='lines'  # Supprime les gros points
))

# Configuration du layout
fig.update_layout(
    polar=dict(
        radialaxis=dict(
            visible=True,
            range=[0, 10]  # Plage des valeurs
        )
    ),
    showlegend=True,
    legend=dict(
        x=1.1,  # Position à droite
        y=1.1
    ),
    width=800,  # Largeur de l'image
    height=800  # Hauteur de l'image
)

# Affichage
fig.show()
```
